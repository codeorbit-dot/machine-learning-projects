## 1. Time & Space Complexity

## 2. Arrays

```python
arr=[10,20,30]
arr.append(40)  # Add a new item to the end of the array
print(arr)
print(arr[1])  # Access the second element
```

## 3. Linked List

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

head=Node(10)  # Create the first node
head.next=Node(20)  # Link the second node
head.next.next=Node(30)  # Link the third node
cur=head
while cur:
    print(cur.data)  # Print data while traversing the linked list
    cur=cur.next
```

## 4. Stack

```python
stack=[]
stack.append(10)  # Push item onto stack
stack.append(20)
stack.append(30)
stack.append(40)
stack.append(50)
print(stack)
print(stack.pop(2))  # Remove item at index 2 from the stack
```

## 5. Queue

```python
from collections import deque
q=deque([10,20,30])  # Create a queue using deque
print(q.popleft())  # Remove and print the front item
```

## 6. Dictionary / Hash Map

```python
text='banana'
freq={}
for ch in text:
    freq[ch]=freq.get(ch,0)+1
print(freq)
```

## 7. Binary Tree

```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

root=TreeNode(10)  # Create the root node
root.left=TreeNode(5)  # Create left child
root.right=TreeNode(15)  # Create right child
print(root.data)  # Print the root value
```

## 8. Linear Search

```python
arr=[10,20,30,40]
target=30
for i,v in enumerate(arr):
    if v==target:
        print('Found at',i)  # Print the index where the target is found
```

## 9. Binary Search

```python
arr=[10,20,30,40,50]
l,r=0,len(arr)-1
target=40
while l<=r:
    m=(l+r)//2  # Find the middle element
    if arr[m]==target:
        print('Found')  # Target found
        break
    elif arr[m]<target:
        l=m+1  # Search in the right half
    else:
        r=m-1  # Search in the left half
```

## 10. Bubble Sort

```python
arr=[5,3,8,1]
for i in range(len(arr)):
    for j in range(len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]  # Swap adjacent elements if out of order
print(arr)  # Print the sorted array
```

## 11. Selection Sort

```python
arr=input().split()
print("Before sorting:", arr)
for i in range(len(arr)):
    mi=i
    for j in range(i+1,len(arr)):
        if arr[j]<arr[mi]:
            mi=j  # Track the smallest element index
    arr[i],arr[mi]=arr[mi],arr[i]  # Swap current element with the minimum
print("After sorting:", arr)
```

## 12. Insertion Sort

```python
arr=[12,11,13,5,6]
for i in range(1,len(arr)):
    key=arr[i]  # Current element to insert
    j=i-1
    while j>=0 and arr[j]>key:
        arr[j+1]=arr[j]  # Shift larger elements right
        j-=1
    arr[j+1]=key  # Place the key in the correct position
print(arr)
```

## 13. Merge Sort

```python
def merge_sort(a):
    if len(a)<=1:return a
    m=len(a)//2
    l=merge_sort(a[:m])  # Sort left half
    r=merge_sort(a[m:])  # Sort right half
    res=[]
    i=j=0
    while i<len(l) and j<len(r):
        if l[i]<r[j]: res.append(l[i]); i+=1
        else: res.append(r[j]); j+=1
    return res+l[i:]+r[j:]  # Combine the sorted halves

print(merge_sort([38,27,43,3]))
```

## 14. Quick Sort

```python
def quick_sort(arr):
    if len(arr)<=1:return arr
    p=arr[len(arr)//2]  # Choose middle element as pivot
    left=[x for x in arr if x<p]  # Elements smaller than pivot
    mid=[x for x in arr if x==p]  # Elements equal to pivot
    right=[x for x in arr if x>p]  # Elements greater than pivot
    return quick_sort(left)+mid+quick_sort(right)

print(quick_sort([10,7,8,9,1,5]))
```

## 15. Recursion - Factorial & Fibonacci

```python
def factorial(n):
    return 1 if n==1 else n*factorial(n-1)  # Recursive factorial calculation

print(factorial(5))

def fibonacci(n):
    if n<=1:return n  # Base case for Fibonacci sequence
    return fibonacci(n-1)+fibonacci(n-2)  # Recursive Fibonacci calculation

print(fibonacci(7))
```
