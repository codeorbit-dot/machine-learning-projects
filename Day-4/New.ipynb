```python

import pandas as pd
import numpy as np

students = {
    "Name": ["Tarun", "Rahul", "Anjali", "Priya", "Kiran"],
    "Math": [85, 78, 92, 88, 75],
    "Science": [90, 80, 95, 85, 70],
    "English": [82, 76, 89, 91, 72]
}

df = pd.DataFrame(students)

print("Student Dataset")
print(df)

# Average marks
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

print("\nAverage Marks")
print(df[["Name", "Average"]])

# Highest scorer
top_student = df.loc[df["Average"].idxmax()]

print("\nHighest Scorer")
print(top_student)

# Variance
print("\nVariance")
print(df[["Math", "Science", "English"]].var())

# Standard Deviation
print("\nStandard Deviation")
print(df[["Math", "Science", "English"]].std())

# Correlation Matrix
print("\nCorrelation Matrix")
print(df[["Math", "Science", "English"]].corr())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Study_Hours": [1,2,3,4,5,6,7,8],
    "Marks": [35,40,50,55,65,70,80,90]
})

plt.figure(figsize=(8,5))
plt.scatter(df["Study_Hours"], df["Marks"])

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```
