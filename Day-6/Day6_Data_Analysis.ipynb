## NumPy Arrays

```python
# numpy is a powerful library for numerical computing in Python. 
# It provides support for large multi-dimensional arrays and matrices, 
# along with a collection of mathematical functions to operate on these arrays.
import numpy as np

arr1 = np.array([10, 20, 30, 40, 50])
print(arr1)

matrix = np.array([[1,2,3],[4,5,6]])
print(matrix)

print("Shape:", matrix.shape)
print("Dimensions:", matrix.ndim)
print("Size:", matrix.size)
print("Data Type:", matrix.dtype)

print(np.zeros((2,3)))
print(np.ones((3,3)))
print(np.arange(1,11))
print(np.linspace(0,1,5))
```

## Broadcasting

```python
#Filtering means selecting only required rows
import numpy as np

arr = np.array([10,20,30])
print(arr + 5)

a = np.array([1,2,3])
b = np.array([10,20,30])
print(a + b)

matrix = np.array([[1,2,3],[4,5,6]])
row = np.array([10,20,30])
print(matrix + row)
```

## Pandas DataFrame

## Filtering Data

```python
#Filtering means selecting only required rows.
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C","D"],
    "Age":[20,25,30,35]
})

print(df[df["Age"] > 25])

result = df[(df["Age"] > 20) & (df["Age"] < 35)]
print(result)

print(df[df["Name"].isin(["A","C"])])
```

## GroupBy

```python
#Used for summarizing data.
import pandas as pd

df = pd.DataFrame({
    "Department":["IT","HR","IT","HR","Finance"],
    "Salary":[50000,40000,60000,45000,70000]
})

print(df.groupby("Department")["Salary"].mean())

print(df.groupby("Department")["Salary"].agg(["mean","max","min","sum"]))
```

## Missing Value Handling

```python
#Real-world datasets often contain missing values.
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["A","B","C","D"],
    "Age":[20,np.nan,30,np.nan]
})

print(df.isnull().sum())

df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)

print(df.dropna())
```

## Data Preprocessing

```python
#Data preprocessing prepares data for Machine Learning.
import pandas as pd

df = pd.DataFrame({
    "Name":["A","A","B","C"],
    "Age":[20,20,25,30]
})

print(df.drop_duplicates())

df.rename(columns={"Age":"Employee_Age"}, inplace=True)
print(df)

df["Employee_Age"] = df["Employee_Age"].astype(float)
print(df.dtypes)
```

## Feature Engineering

```python
#Creating new useful columns from existing columns.
import pandas as pd

df = pd.DataFrame({
    "Math":[80,90,75],
    "Science":[85,95,70]
})

df["Total"] = df["Math"] + df["Science"]
df["Average"] = df["Total"] / 2

print(df)
```

## Mini Project

```python
import pandas as pd
import numpy as np

data = {
    "Name":["A","B","C","D","E"],
    "Department":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,np.nan,60000,70000,45000],
    "Experience":[2,4,5,8,3]
}

df = pd.DataFrame(data)

df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

df["Salary_Per_Year"] = df["Salary"] / df["Experience"]

high_salary = df[df["Salary"] > 50000]

department_avg = df.groupby("Department")["Salary"].mean()

print(df)
print(high_salary)
print(department_avg)
```
