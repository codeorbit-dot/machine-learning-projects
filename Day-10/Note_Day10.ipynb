Important Docker Commands

# Check version
docker --version

# Pull image
docker pull ubuntu

# View images
docker images

# Run container
docker run ubuntu

# Run interactive container
docker run -it ubuntu bash

# Running containers
docker ps

# All containers
docker ps -a

# Stop container
docker stop <container_id>

# Remove container
docker rm <container_id>
