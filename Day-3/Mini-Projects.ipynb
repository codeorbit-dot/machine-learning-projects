```python
#!/usr/bin/env python3
#import necessary libraries
import requests
import json
import logging
import time
from abc import ABC, abstractmethod
```

```python
# LOGGING
logging.basicConfig(
    filename="github_analyzer.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
```

```python
# DECORATOR
def execution_timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()

        print(f"\nExecution Time: {end-start:.2f} seconds")
        return result

    return wrapper
```

```python
# ABSTRACTION
class Repository(ABC):

    def __init__(self, name, stars):
        self._name = name
        self._stars = stars

    @abstractmethod
    def show_info(self):
        pass
```

```python
# INHERITANCE
class PythonRepo(Repository):

    def show_info(self):
        return f"Python Repo: {self._name} | Stars: {self._stars}"


class OtherRepo(Repository):

    def show_info(self):
        return f"Other Repo: {self._name} | Stars: {self._stars}"

```

```python
# ENCAPSULATION

class User:

    def __init__(self, username):
        self.__username = username

    @property
    def username(self):
        return self.__username
```

```python
# ITERATOR

class RepoIterator:

    def __init__(self, repos):
        self.repos = repos
        self.index = 0

    def __iter__(self):
        self.index = 0
        return self

    def __next__(self):

        if self.index >= len(self.repos):
            raise StopIteration

        item = self.repos[self.index]
        self.index += 1

        return item
```

```python
# GENERATOR

def top_repositories(repositories, min_stars):

    for repo in repositories:
        if repo._stars >= min_stars:
            yield repo

```

```python
# CONTEXT MANAGER

class ReportGenerator:

    def __init__(self, filename):
        self.filename = filename

    def __enter__(self):
        self.file = open(self.filename, "w", encoding="utf-8")
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.file.close()

```

```python
# API CONSUMPTION

@execution_timer
def fetch_repositories(topic):

    url = f"https://api.github.com/search/repositories?q={topic}&sort=stars"

    response = requests.get(url, timeout=10)

    response.raise_for_status()

    data = response.json()

    logging.info("GitHub API called")

    return data["items"][:20]
```

```python
# JSON HANDLING

def save_json(data):

    with open("repos.json", "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4)

    logging.info("JSON saved")


def load_json():

    with open("repos.json", "r", encoding="utf-8") as file:
        return json.load(file)

```

```python
# MAIN

def main():

    print("\n=== GITHUB REPOSITORY ANALYZER ===\n")

    user = User("Developer")

    print("User:", user.username)

    topic = input("Enter topic (python, ai, machine-learning): ")

    repositories = fetch_repositories(topic)

    save_json(repositories)

    repositories = load_json()

    repo_objects = []

    for repo in repositories:

        if repo["language"] == "Python":
            repo_objects.append(
                PythonRepo(
                    repo["name"],
                    repo["stargazers_count"]
                )
            )
        else:
            repo_objects.append(
                OtherRepo(
                    repo["name"],
                    repo["stargazers_count"]
                )
            )

    print("\n--- POLYMORPHISM DEMO ---")

    for repo in repo_objects[:5]:
        print(repo.show_info())

    print("\n--- ITERATOR DEMO ---")

    iterator = RepoIterator(repo_objects)

    for repo in iterator:
        print(repo.show_info())

    print("\n--- GENERATOR DEMO ---")

    min_stars = int(input("\nMinimum stars: "))

    for repo in top_repositories(repo_objects, min_stars):
        print(repo.show_info())

    print("\n--- REPORT GENERATION ---")

    with ReportGenerator("github_report.txt") as report:

        report.write("GITHUB REPOSITORY REPORT\n\n")

        for repo in repo_objects:
            report.write(repo.show_info() + "\n")

    print("Report Saved Successfully")

    logging.info("Program completed")


if __name__ == "__main__":
    main()
```
