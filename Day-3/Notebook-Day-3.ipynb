```python
#!/usr/bin/env python3
#simple class to represent a student with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("Tarun", 95)
s1.display()
```

```python
#simple class to represent an employee with name and work
#Inheritance
class Employee:
    def __init__(self, name):
        self.name = name

    def work(self):
        print(self.name, "is working")

class Developer(Employee):
    def code(self):
        print(self.name, "is coding")

d = Developer("Tarun")

d.work()
d.code()
```

```python
#Polymorphism
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

```python
#Encapsulation
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def show_balance(self):
        print("Balance:", self.__balance)

acc = BankAccount()

acc.deposit(5000)
acc.show_balance()
```

```python
#Abstraction
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")

c = Car()
c.start()
```

```python
# Decorators
def decorator(func):

    def wrapper():
        print("Before Function")
        func()
        print("After Function")

    return wrapper

@decorator
def greet():
    print("Hello")

greet()
```

```python
# Context Managers
with open("sample.txt", "w") as file:
    file.write("Hello Python")
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
class FileManager:

    def __enter__(self):
        print("File Opened")
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print("File Closed")

with FileManager():
    print("Working")
```

```python
import json

class Employee:

    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def to_dict(self):
        return {
            "id": self.emp_id,
            "name": self.name,
            "salary": self.salary
        }

employees = []

while True:

    print("\n1.Add")
    print("2.View")
    print("3.Save")
    print("4.Exit")

    choice = input("Enter Choice: ")

    if choice == "1":

        emp_id = input("ID: ")
        name = input("Name: ")
        salary = float(input("Salary: "))

        employees.append(Employee(emp_id, name, salary))

    elif choice == "2":

        for emp in employees:
            print(emp.emp_id, emp.name, emp.salary)

    elif choice == "3":

        data = []

        for emp in employees:
            data.append(emp.to_dict())

        with open("employees.json", "w") as file:
            json.dump(data, file, indent=4)

        print("Saved Successfully")

    elif choice == "4":
        break
```
