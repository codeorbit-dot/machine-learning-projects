```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [120, 150, 180, 200, 250]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

courses = ["Python", "SQL", "ML", "AI"]
students = [50, 40, 30, 25]

plt.bar(courses, students)
plt.title("Students per Course")
plt.xlabel("Course")
plt.ylabel("Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["Python", "SQL", "ML", "AI"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.title("Course Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [55,60,65,70,72,75,80,82,85,88,90,95]

plt.hist(marks, bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6,7,8]
marks = [35,40,50,55,65,70,80,90]

plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
