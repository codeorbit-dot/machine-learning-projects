```python
n = int(input("How many terms? "))
a, b = 0, 1

print("Fibonacci series:", end=" ")

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python

num = int(input("Enter number: "))
if num > 1:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```
